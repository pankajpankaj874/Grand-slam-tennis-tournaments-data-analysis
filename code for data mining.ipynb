```python
# question 1, Dataset 1 (AusOpen-men-2013.csv)

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("AusOpen-men-2013.csv")

df1=df[df['ST1.1']>df['ST1.2'] ]
n1 = len(df1)
df2 = df1[df1['Result']==1]
n2 = len(df2)

P1 = n2/n1
print("Probability of a player winning the match if he won the  first set :",P1)

df3 = df1[df1['ST1.1']==1+df1['ST1.2']  ]
n3 = len(df3)
df4 = df3[df3['Result']==1]
n4 = len(df4)

P2 = n4/n3
print("Probability of a player winning the match if he won the  first set by only 1 point:",P2)

Percent1 = P1*100
Percent2 = P2*100

labels1 = ['won=Percent1','lost']
data1 = [Percent1,100-Percent1]
labels2 = ['won=Percent2','lost']
data2 = [Percent2,100-Percent2]

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.pie(data1,labels=labels1,autopct='%1.1f%%')
ax1.set_title("Player won after winning first set")
ax2.pie(data2,labels=labels2,autopct='%1.1f%%')
ax2.set_title("Player won after winning first set by only 1 point")
```

```python
# question 2, Dataset 2 (AusOpen-women-2013.csv)
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('AusOpen-women-2013.csv')
X = df[['UFE.1']]
y = df['Result']

model = LinearRegression()
model.fit(X, y)

predicted = model.predict(X)

print('Coefficients: ', model.coef_)
print("Intercept: ", model.intercept_)

plt.plot(X, model.predict(X), color='blue', linewidth=3)
plt.xlabel('Unforced Errors committed by player 1')
plt.ylabel('Result')
plt.show()
```

```python
# question 4, Dataset 4 (FrenchOpen-women-2013.csv)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('FrenchOpen-women-2013.csv')
df1 = df[df['Round']==1]
n1= len(df1)
S1 = df1['UFE.1'].sum()+df1['UFE.2'].sum()
avg1 = S1/n1

df2 = df[df['Round']==2]
n2= len(df2)
S2 = df2['UFE.1'].sum()+df2['UFE.2'].sum()
avg2 = S2/n2

df3 = df[df['Round']==3]
n3= len(df3)
S3 = df3['UFE.1'].sum()+df3['UFE.2'].sum()
avg3 = S3/n3

df4 = df[df['Round']==2]
n4= len(df4)
S4 = df4['UFE.1'].sum()+df4['UFE.2'].sum()
avg4 = S4/n4

df5 = df[df['Round']==5]
n5= len(df5)
S5 = df5['UFE.1'].sum()+df5['UFE.2'].sum()
avg5 = S5/n5

df6 = df[df['Round']==6]
n6= len(df6)
S6 = df6['UFE.1'].sum()+df6['UFE.2'].sum()
avg6 = S6/n6

df7 = df[df['Round']==7]
n7= len(df7)
S7 = df7['UFE.1'].sum()+df7['UFE.2'].sum()
avg7 = S7/n7

avg_unforced_errors = [avg1,avg2,avg3,avg4,avg5,avg6,avg7]
rounds = ['Round 1', 'Round 2', 'Round 3', 'Round 4', 'Quarterfinals', 'Semifinals', 'Finals']
print("Average unforced errors roundwise:",avg_unforced_errors)
plt.plot(rounds, avg_unforced_errors)

plt.xlabel("Round")
plt.ylabel("Average Unforced Errors")
plt.title("Average Unforced Errors per Round")

plt.show()
```

```python
# question 5, Dataset 5 (USOpen-men-2013.csv)
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("USOpen-men-2013.csv")
corr = data['BPC.1'].corr(data['TPW.1'])

plt.scatter(data['BPC.1'][data['Result']==1], data['TPW.1'][data['Result']==1], color='green', label='Win')
plt.scatter(data['BPC.1'][data['Result']==0], data['TPW.1'][data['Result']==0], color='red', label='Loss')

plt.xlabel('Break Points Created by player 1')
plt.ylabel('Total points won by player 1')
plt.title('Relationship between Break Points Created and total point won (Correlation: {})'.format(round(corr, 2)))
plt.legend()
plt.show()
```

```python
# question 8, Dataset 8 (Wimbledon-women-2013.csv)
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Wimbledon-women-2013.csv")

df1 = data[data['BPC.1']>8]
df2 = df1[df1['Result']==1]
p = len(df2)/len(df1)
print("Probability that if a player has more than 8 breakpoints then he will win the match :",p)
corr = data['BPC.1'].corr(data['UFE.1'])

data_df = pd.DataFrame({'BPC.1': [p*100, 100-(p*100)],
                        'UFE.1': [0, 0]})

label = ["win if breaking pts create>8","losing"]

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.scatter(data['BPC.1'],data['UFE.1'])
ax1.set_title('correlation between break points created and unforced errors')

ax2.pie(data_df['BPC.1'], labels=label,autopct='%1.1f%%')

plt.show()
```
